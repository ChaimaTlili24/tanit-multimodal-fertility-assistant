## End-to-end multimodal pipeline test

```python
from rag.rag import retrieve_context
from vlm import analyze_image, analyze_pdf

# Simulated user text (no audio, no real file here)
user_text = "I am 30, my AMH is 2.1 ng/mL and my cycles are irregular."
image_summary = analyze_image(None, user_text)   # no image provided
pdf_summary = analyze_pdf(None, user_text)       # no PDF provided

extra_parts = []
if image_summary:
    extra_parts.append(f"Image: {image_summary}")
if pdf_summary:
    extra_parts.append(f"PDF: {pdf_summary}")

multi_context = " | ".join(extra_parts) or "[no multimodal context]"
rag_context = retrieve_context(user_text)

print("=== User text ===")
print(user_text)
print("\n=== RAG medical context (stub) ===")
print(rag_context)
print("\n=== Multimodal context (stub) ===")
print(multi_context)
```
