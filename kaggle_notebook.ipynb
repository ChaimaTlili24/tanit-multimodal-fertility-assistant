## End-to-end multimodal pipeline test

```python
from rag import retrieve_context
from vlm import analyze_image, analyze_pdf

user_text = "J'ai 30 ans, AMH à 2.1 ng/mL et cycles irréguliers."
image_summary = analyze_image(None, user_text)
pdf_summary = analyze_pdf(None, user_text)

extra_parts = []
if image_summary:
    extra_parts.append(f"Image: {image_summary}")
if pdf_summary:
    extra_parts.append(f"PDF: {pdf_summary}")

extra_context = " | ".join(extra_parts)
rag_context = retrieve_context(user_text)

print("=== Question utilisateur ===")
print(user_text)
print("\n=== Contexte RAG ===")
print(rag_context)
print("\n=== Contexte multimodal (simulé) ===")
print(extra_context or "[aucun]")
```
